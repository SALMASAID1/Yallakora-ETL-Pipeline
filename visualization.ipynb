```python
import pandas as pd;
df=pd.read_csv('LastData.csv')
print(df)
```

```python
import matplotlib.pyplot as plt;
import seaborn as sns;
```

```python
# Matches per Championship

championship_counts = df['Championship Title'].value_counts()

# championship_counts.plot(kind='pie')
plt.pie(championship_counts,labels=championship_counts.index, autopct='%1.1f%%')

plt.title('Number of Matches per Championship')
plt.show()
```

```python
bestscore={ "score":[],"Winner":[] }
```

```python
import re
for i in range (len(df['Score'])):
 
   print(df['Score'][i]) 
   number_scor = re.findall("[0-9]+",df['Score'][i]) 
   if (int(number_scor[0])>int(number_scor[1])):
      print("winner is Team 1 ---> with score =", int(number_scor[0]))
      bestscore['score'].append(int(number_scor[0]))
      bestscore['Winner'].append(df['Team 1'][i])

   elif(int(number_scor[0])<int(number_scor[1])):
      print("winner is Team 2 ---> with score =", int(number_scor[1]))
      bestscore['score'].append(int(number_scor[1]))
      bestscore['Winner'].append(df['Team 2'][i])
   else:
      print("No winner ")
print(bestscore)
```

```python
plt.figure(figsize=(10, 6))
Y=bestscore['score']
X=bestscore['Winner']
plt.xticks(rotation=80) 
plt.bar(X,Y)
plt.title("Distribution Winners")
plt.xlabel("winner teams")
plt.ylabel("score")
plt.show()
```
